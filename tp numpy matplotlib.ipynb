```python
import pandas as pd 
import matplotlib.pyplot as plt 
  
df = pd.DataFrame({ 
    'Object': ['Bulb', 'Lamp', 'Table', 'Pen', 'Notebook'], 
    'Price': [45, 38, 90, 60, 40] 
}) 
  
plt.pie(df["Price"], labels=df["Object"]) 
plt.show() 
```

```python
# import package
import turtle
  
# set speed to slowest for
# better understandings
turtle.speed(1)
  
# motion
turtle.forward(100)
turtle.right(90)
  
# hide the turtle
turtle.hideturtle()
  
# motion
turtle.forward(100)
turtle.right(90)
turtle.forward(100)
turtle.right(90)
  
# show the turtle
turtle.showturtle()
  
# motion
turtle.forward(100)
turtle.right(90)
```
